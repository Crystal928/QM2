```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pylab
from datetime import datetime

%matplotlib inline
pylab.rcParams['figure.figsize'] = (10, 8)

aqi_data=pd.read_csv(r'./data/wk1/california_aqi.csv')
aqi_data
```

```python
#1、Sort that dataframe to figure out which day had the second worst AQI.
aqi_avg_data= aqi_data.groupby('Date')['AQI'].mean()
dict_date = {'Date':aqi_avg_data.index,'AQI':aqi_avg_data.values}
aqi_avg_data = pd.DataFrame(dict_date)

aqi_avg_data.sort_values(by='AQI', inplace=True, ascending=False)
aqi_avg_data
```

From the above data, it can be seen 9/13/20 has the second worst AQI.

```python
#2、Plug that date into the satellite_plot() function to visualize the \
# corresponding satellite image. If you’ve done things correctly so far, \
# you should see a big plume of smoke emanating from the mountains of california \
# spreading northwards.

from ipyleaflet import Map, Marker, basemaps, basemap_to_tiles,Circle
from ipywidgets import HTML

aqi_data['Date']=pd.to_datetime(aqi_data['Date'])

# scrape the table of AQI values and corresponding colors 
# save it as a dataframe called colors
colors= pd.read_html('https://www.airnow.gov/aqi/aqi-basics/')[0]

# create a numerical column for AQI values by splitting the test in the "values of index" column. 
# pull out the first string, and convert it to integer
colors['aqi']=colors['Values of Index'].str.split(' ').str[0].astype(int)

aqi_colors=matplotlib.colors.LinearSegmentedColormap.from_list(colors['aqi'],colors['Daily AQI Color'])


def satellite_plot(date):
  
  ymd=datetime.strptime(date, '%d-%m-%Y').strftime('%Y-%m-%d')

  satellite_map = Map(
    basemap=basemap_to_tiles(
      basemaps.NASAGIBS.ModisTerraTrueColorCR, ymd
    ),
    center=(36.77, -119.41),
    zoom=6,
  )

  satellite_map.layout.height = '700px'

  one_day=aqi_data[aqi_data['Date']==date]
  
  for index, row in one_day.iterrows():
    color=matplotlib.colors.rgb2hex(aqi_colors(row['AQI']))
    point=Circle(location=(row['latitude'],row['longitude']), color=color)
    point.popup = HTML(str(row['Site Name']))    
    satellite_map.add_layer(point)
  return satellite_map
```

```python
satellite_plot('13-09-2020')
```

```python
import pandas as pd 

tweets=pd.read_csv('https://storage.googleapis.com/qm2/wk4/Exxon_tweets_clean.csv')
tweets=tweets.sample(frac=1, random_state=1) # don't edit this line, i'm shuffling the data for you
tweets
```

```python
#1、create a dataframe called “sample” which contains the first 3000 tweets from the “tweets” dataframe.

sample = tweets[:3000]
sample
```

```python
# 2、Using a lambda function, .apply(lambda x: nlp(x)._.blob.polarity), 
# create a column in the sample dataframe that contains the polarity of each tweet.
import spacy
import pylab
from IPython.core.display import display, HTML
pylab.rcParams['figure.figsize'] = (10., 8.)
from spacytextblob.spacytextblob import SpacyTextBlob

nlp = spacy.load("en_core_web_sm")
nlp.add_pipe('spacytextblob')

sample['polarity'] = sample['text'].apply(lambda x: nlp(x)._.blob.polarity)

sample
```

```python
#3、Create a column that contains the subjectivity score for each tweet.
sample['subjectivity'] = sample['text'].apply(lambda x: nlp(x)._.blob.subjectivity)

sample
```

```python
# 4、Filter the dataframe to keep only the tweets that are subjective (subjectivity score > 0.5) 
# and tweets that have negative sentiment (polarity score < 0).

sample_filter = sample[(sample['subjectivity'] > 0.5) & (sample['polarity'] < 0.0)]
sample_filter
```

```python
# Question: which twitter user hates Exxon the most (lowest total sentiment)? 
# Make note of their ID.

sample_hates_Exxon_the_most = sample[(sample['subjectivity'] > 0.75) & (sample['polarity'] < -0.75)]
res_list = sample_hates_Exxon_the_most['author_id'].values.tolist()

print(f"The IDs of twitter user hates Exxon the most (lowest total sentiment): \n {res_list}")
```

```python
# 1、Create a new dataframe which restricts the census data to just white 
# men who have completed highschool (and only highschool).

white_men_completed_highschool_data =  df[(df['sex'] == 1) & ((df['race'] == 1) | (df['race'] == 3)) & (df['sch'] >= 13)]
white_men_completed_highschool_data
```

```python
#2、Do the same for hispanic women with a Bachelor’s degree (and only a Bachelor’s).

hispanic_women_Bachelor_degree_data = df[(df['sex'] == 2) & (df['race'] == 3) & (df['sch'] == 16)]
hispanic_women_Bachelor_degree_data.columns
```

```python
#3、Use the two_hist() function above to plot these two income groups against each other.
def two_hist(groups,group_labs,xlab, title): # define a function called two_hist that takes a list of groups, a list of group labels, an x-axis label, and a title as arguments

        plt.figure(figsize=(15,6)) # set the figure size

        it=-1 # create a counter variable called "it" and set it equal to -1
        for var in groups: # loop through each group in the list of groups
            it+=1 # increase the iterator by 1
            sample_size=1000 # set the sample size equal to 1000
            sample_means=[] # create an empty list to store sample means
            iterations=10000 # set the number of iterations equal to 10,000

            for i in range(0,iterations): # loop through the number of iterations
                sample=var.sample(sample_size, replace=True) # draw a sample of the size specified by the "sample_size" parameter
                sample_mean=sample.mean() # calculate the mean of the sample
                sample_means.append(sample_mean) # append the sample mean to the list of sample means
            
            plt.hist(sample_means, bins=int(iterations/300),edgecolor='white',density=True, label=group_labs[it])  # plot a histogram of the sample means
            mu, se = norm.fit(sample_means) # fit a normal distribution to the data
            xmin, xmax = plt.xlim() # set the x-axis limits
            x = np.linspace(xmin, xmax, 100) # create a range of values from the minimum to the maximum x-axis value
            p = norm.pdf(x, mu, se) # calculate the probability density function for the normal distribution

            mean=var.mean() # calculate the mean of the variable
            lower_ci=mu-se*1.96 # calculate the lower bound of the 95% confidence interval
            upper_ci=mu+se*1.96 # calculate the upper bound of the 95% confidence interval

            plt.plot(x, p, 'k', linewidth=2) # plot the normal distribution
            plt.xlabel(xlab) # add an x-axis label
            plt.title(title) # add a title
            plt.axvline(mean, color='green', linestyle='solid', linewidth=3) # add a vertical line at the mean of the variable
            plt.axvline(lower_ci, color='black', linestyle='dashed', linewidth=1.5) # add a vertical line at the mean minus 2 standard deviations
            plt.axvline(upper_ci, color='black', linestyle='dashed', linewidth=1.5) # add a vertical line at the mean plus 2 standard deviations
            plt.legend() # add a legend
            
        plt.show()  # show the plot 


two_hist([white_men_completed_highschool_data['incwage'], hispanic_women_Bachelor_degree_data['incwage']],['white Men completed highschool','hispanic women with a Bachelor’s degree'],'Income ($, thousands)', "Income Sample Means")
```
