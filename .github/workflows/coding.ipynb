```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = r'D:/workspace/学习/python-开发-20241110/heat_map.csv'
data = pd.read_csv(file_path)
heatmap_data = data.sort_values(by='rate', ascending=False)
heatmap_data = heatmap_data.pivot(index='country', columns='year', values='rate')

plt.figure(figsize=(14, 10))
sns.heatmap(heatmap_data, cmap='coolwarm', cbar_kws={'label': 'malaria death rate'})
plt.title('Annual malaria death rate (1990-2021) for African Countries')
plt.xlabel('Year')
plt.ylabel('Country')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="white")

file_path = r'D:/workspace/学习/python-开发-20241110/line_chart.csv'
data = pd.read_csv(file_path)


plt.figure(figsize=(10, 8))
# Plot the responses for different events and regions
sns.lineplot(x="year", y="rate",hue="country", err_style="band", data=data)
plt.title('Annual malaria death rate (1990-2021) for Some African Countries')
plt.xlabel('Year')
plt.ylabel('malaria death rate')
plt.show()
```

```python
from pyecharts.charts import Geo
from pyecharts.datasets import register_url
from pyecharts import options as opts
from pyecharts.globals import ChartType, SymbolType
from pyecharts.charts import Map
import pandas as pd

map_1990_data = pd.read_csv(r'D:/workspace/学习/python-开发-20241110/1990_map.csv')
rates_list = map_1990_data['rate'].values.tolist()
countries_list = map_1990_data['country'].values.tolist()
data_res = []
for i in range(len(countries_list)):
    data_res.append((countries_list[i], rates_list[i]))

low, high = 0.0, 260.0             
(
    Map()
    .add("", data_res, maptype="world", is_map_symbol_show=False)
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(visualmap_opts=opts.VisualMapOpts(max_=high, min_=low),
                     title_opts=opts.TitleOpts(
                         title="1990  malaria death rate for African Countries")
                     )
    .render_notebook()
)
```

```python
from pyecharts.charts import Geo
from pyecharts.datasets import register_url
from pyecharts import options as opts
from pyecharts.globals import ChartType, SymbolType
from pyecharts.charts import Map
import pandas as pd

map_average_data = pd.read_csv(r'D:/workspace/学习/python-开发-20241110/heat_map.csv')
result_mean = map_average_data[['country', 'rate']].groupby('country').mean()
# countries_list = map_2021_data['country'].values.tolist()
countries_list = result_mean.index.tolist()
rates_list = result_mean['rate'].values.tolist()

data_res = []
for i in range(len(countries_list)):
    data_res.append((countries_list[i], rates_list[i]))

low, high = 0.0, 230.0             
(
    Map()
    .add("", data_res, maptype="world", is_map_symbol_show=False)
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(visualmap_opts=opts.VisualMapOpts(max_=high, min_=low),
                     title_opts=opts.TitleOpts(
                         title="Annual average malaria death rate (1990-2021) for African Countries")
                     )
    .render_notebook()
)
```
